## Mostrar la información del dataset

En este caso, se va a estudiar el comportamiento del atributo **MEDV**, que hace referencia al *valor medio de la vivienda*, usando el número medio de habitaciones.

```python
print(boston.DESCR)
```

## Objetivo:

Utilizando el número medio de habitaciones(**RM**), intentar predecir el valor medio de la vivienda (**MEDV**).

```python
#Se requieren que los datos esten en forma de array, por eso se usa np.array() 

#Columna de valor medio de habitaciones
#[5] sería ala fila 5, pero como se quiere la columna hay que poner [:, 5]
rm = np.array(boston.data[:, 5])
#Columna de valor medio de la vivienda
medv = np.array(boston.target)
```

### Gráfica de la nube de puntos

Donde la *X* va a ser **RM** y la *Y* va a ser **MEDV** usadno *matplot*.

```python
plt.scatter(rm, medv, alpha=0.3) #aplha es la transparencia de los puntos
plt.show()
```

### Fórmula para minimizar el error cuadrático medio (MCO)

$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
#Anadimos columna de 1s para tratar el término independiente
rm_mod = np.array([np.ones(506), rm]).T #Necesitamos la transpuesta
```

```python
#Escribir la formula (@ es multiplicacion matricial)
B = np.linalg.inv(rm_mod.T @ rm_mod) @ rm_mod.T @ medv
```

```python
B
```

### Dibujo de la gráfica

```python
plt.plot([3.5, 9], [B[0]+B[1]*3.5, B[0]+B[1]*9], c="red")
plt.scatter(rm, medv, alpha=0.3) #aplha es la transparencia de los puntos
plt.show()
```
